```python
datos['Churn'].value_counts(dropna=False) # identificar cuantos valores vac铆os hay en la columna 'Churn'
```

```python
datos = datos[datos['Churn'].isin(['Yes', 'No'])].copy()  #Eliminar filas vac铆as en la columna 'Churn'
```

```python
datos['account.Charges.Total'] = pd.to_numeric(datos['account.Charges.Total'], errors='coerce')
datos['account.Charges.Total'] = datos['account.Charges.Total'].fillna(0)
datos['account.Charges.Total'].dtype
```

```python
 #Lista con valores 'Yes'/'No' en columnas
columnas_b =[
    'Churn',
    'account.PaperlessBilling',
    'customer.Partner',
    'customer.Dependents',
    'internet.TechSupport'
]
```

```python
for col in columnas_b:
    datos[col] = datos[col].map({'Yes': True, 'No': False})
datos.info()
```
